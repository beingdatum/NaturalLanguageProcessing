```python
import re
def generate_ngrams(text,n): 
    # split sentences into tokens
    tokens=re.split("\\s+",text)
    ngrams=[] 
    # collect the n-grams
    for i in range(len(tokens)-n+1):
    	temp=[tokens[j] for j in range(i,i+n)]
    	ngrams.append(" ".join(temp)) 
    return ngrams

generate_ngrams('This is sparta', 2)
```
